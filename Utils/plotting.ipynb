```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.linear_model import LinearRegression

```

```python
np.random.seed(42)
X1 = np.random.rand(100) * 10
X2 = np.random.rand(100) * 10
Y = 2* X1 + 3*X2 + 5 + np.random.randn(100) * 2

df = pd.DataFrame({'X1':X1, 'X2': X2, 'Y':Y})

model = LinearRegression()
model.fit(df[['X1', 'X2']],df['Y'])

```

```python
x1_range = np.linspace(df['X1'].min(), df['X1'].max(),10)
x2_range = np.linspace(df['X2'].min(), df['X2'].max(), 10)
x1_grid, x2_grid = np.meshgrid(x1_range,x2_range)

y_plane = model.coef_[0]* x1_grid + model.coef_[1] * x2_grid + model.intercept_
fig = px.scatter_3d(df, x= 'X1', y = 'X2' , z  = 'Y', title = '3D scatter plot')

fig.add_trace(px.surface(x = x1_grid, y= x2_grid, z = y_plane).data[0])
fig.update_layout(scene=dict(
                    xaxis_title='X1',
                    yaxis_title='X2',
                    zaxis_title='Y'
                ))

# Show the plot
fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random data
np.random.seed(0)
num_cars = 100
hp = np.random.randint(100, 400, num_cars)
mpg = np.random.uniform(10, 50, num_cars)
engine_displacement = np.random.randint(2000, 6000, num_cars)
cylinders = np.random.choice([4, 6, 8, 12], num_cars)

# Step 2: Create a DataFrame
data = {
    'Horsepower': hp,
    'MPG': mpg,
    'Engine Displacement': engine_displacement,
    'Cylinders': cylinders
}
df = pd.DataFrame(data)

# Step 3: Plot a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Horsepower'], df['MPG'], s=df['Engine Displacement']/100, c=df['Cylinders'], cmap='viridis', alpha=0.7)
plt.colorbar(label='Number of Cylinders')
plt.xlabel('Horsepower')
plt.ylabel('Fuel Efficiency (MPG)')
plt.title('Car Engine Performance')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
data = {
    'Price': [300000, 350000, 320000, 400000, 380000, 360000, 450000, 410000, 390000, 500000],
    'House Type': ['Apartment', 'Townhouse', 'Detached House', 'Apartment', 'Detached House',
                   'Townhouse', 'Detached House', 'Apartment', 'Townhouse', 'Detached House'],
    'City': ['Example City'] * 10
}
df = pd.DataFrame(data)

df
# filter data with house as name 
city_name = 'Example City'
city_df = df[df['City']==city_name]
city_df

fig, axes = plt.subplots(1,3,figsize= (15,5),sharey=True, tight_layout = True)
house_types = city_df['House Type'].unique()

for i, hourse_type in enumerate(house_types):

    ax = axes[i]
    sns.histplot(city_df[city_df['House Type'] == hourse_type]['Price'], kde=True, ax = ax, bins = 5 , color = 'C{}'.format(i), label  = hourse_type)
    ax.set_xlabel('Price')
    ax.set_ylabel('Frequency')
    ax.legend()


fig.suptitle('distribution')
plt.show()
```
